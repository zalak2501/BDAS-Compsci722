```python
### 3Merging datasets ###

#  GDP+IE+PTE+SE

fte1  = spark.read.csv("fte1.csv",header=True,inferSchema=True) #full-time employment
gwg1 = spark.read.csv("gwg1.csv", header=True,inferSchema=True) #gender wage gap
gnif1 = spark.read.csv("gnif1.csv", header=True,inferSchema=True) #gross national income female
impexp1= spark.read.csv("impexp1.csv", header=True,inferSchema=True) #import-export cost
gnim1 = spark.read.csv("gnim1.csv", header=True,inferSchema=True) #gross national income male
grad= spark.read.csv("My datasets/Graduation_rate.csv", header=True,inferSchema=True) #graduation rates of men and women
inequality1= spark.read.csv("inequality1.csv", header=True,inferSchema=True) #gender inequality measures
gdp1 = spark.read.csv("gdp1.csv", header=True,inferSchema=True) #gross domestic product
pte1 = spark.read.csv("pte1.csv", header=True,inferSchema=True) #part-time employment
se1= spark.read.csv("se1.csv", header=True,inferSchema=True) #self-employment

## steps 3.4 and 3.5


Mergeddata1=gdp1.withColumnRenamed("SUBJECT","GDPsubject").join(impexp1.withColumnRenamed("SUBJECT","Impexp"),on=['LOCATION','TIME'],how='inner')
Mergeddata1.show()
```

```python

#The datasets 'se1' and 'pte1' both has value as a same column and is not descriptive
Mergeddata2=Mergeddata1.join(pte1.withColumnRenamed("SUBJECT","Sex").withColumnRenamed("Value","pte_value"),on=['LOCATION','TIME'],how='inner')
Mergeddata2.show() #here value defines the parttime rate in %
```

```python
Mergeddata3=Mergeddata2.join(se1.withColumnRenamed("SUBJECT","Sex").withColumnRenamed("Value","se_value"),on=['LOCATION','TIME','Sex'],how='inner')
Mergeddata3.show() #both se and pte contains same column names 
#will change it in next step of format the data
```

```python
#Remaining 5 datasets FTE+GWG+INEQUALITY+GNIM+GNIF

#the column names of gwg are changed 

Mergeddata4=fte1.join(gwg1.withColumnRenamed("2011","2011gwg").withColumnRenamed("2012","2012gwg").withColumnRenamed("2013","2013gwg").withColumnRenamed("2014","2014gwg").withColumnRenamed("2015","2015gwg").withColumnRenamed("2016","2016gwg"),on=['Country'],how='inner')
Mergeddata4.show()

#the gwg has column names which are not quite descriptive-> will change in step 3.5

             
            
```

```python
Mergeddata5=Mergeddata4.join(gnif1,on=['Country'],how='inner')
Mergeddata5.show()
```

```python
Mergeddata6=Mergeddata5.join(gnim1,on=['Country'],how='inner')
Mergeddata6.show()
```

```python
Mergeddata7=Mergeddata6.join(inequality1,on=['Country'],how='inner').drop("_c0","HDI Rank(2015)") 
Mergeddata7.show()
```
