Using Pandas and Matplotlib for getting better visuals-

```python
#######  DATA EXPLORATION  #######

# Full-time employment dataset

import pandas as pd

#using index_col to index data through Country and not any arbitraty numbers like 0,1,2..

fte=pd.read_csv('My datasets/full-time_employ_rate.csv',index_col='Country')
fte.head()

```

```python

#using matplotlib package for better visual

import matplotlib.pyplot as plt 
###Scatter plots with matplotlib
%matplotlib inline  
###If not defined the graph won't be visible
plt.plot(fte['Sex'], fte['2016_fte'],'ro', color='orange')
```

It is clearly seen from above graph that Men fulltime employment rate is higher than Women.

```python
# Gender wage gap dataset
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight') 
gwg=pd.read_csv("My datasets/Gender wage gap.csv")

gwg.head()
```

```python

from matplotlib.pyplot import figure # for changing the size of graph as it was overlapping

figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')

#DEMO-taking few rows for better visual 
gwg1=gwg.head(7) 

plt.plot(gwg1['Country'],gwg1['2015'])
plt.ylabel('Gender Wage Gap')

#Highestwage gap from this is seen in Chile and lowest in Belgium
```

```python
# Inequality dataset

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

inequality=pd.read_csv("My datasets/Gender_inequality.csv",index_col='GII Rank')
inequality.head()
```

```python

from matplotlib.pyplot import figure

plt.style.use('ggplot')
figure(num=None, figsize=(11, 5), dpi=80, facecolor='w', edgecolor='k')

in1=inequality.head(10)

plt.plot(in1['Country'],in1['Percent Representation in Parliament'])
```

```python

in1=inequality.head(10)
figure(num=None, figsize=(20, 5), dpi=80, facecolor='w', edgecolor='k')

plt.bar(in1['Country'],in1['Population with Secondary Education (Female)'],align='center',color='yellowgreen')

#example-united states is having low % in parliament for women but has alomost 95% educated women->gender inequality
```

```python
plt.pie(in1['Labour Force Participation Rate (Female)'],labels=['Norway','Australia','Switzerland','Denmark','Netherlands','Germany',
                                                               'Ireland','US','Canada','NZ'],autopct='%.2f',radius=1.5,startangle=90)
#the values inside chart indicates the fraction of that part in whole pie chart
#almost every country(from 1st 10 selected) has about similar ratio
```
