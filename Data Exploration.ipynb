```python
#######  DATA EXPLORATION  #######

# Full-time employment dataset

import pandas as pd

#using index_col to index data through Country and not any arbitraty numbers like 0,1,2..

fte=pd.read_csv('My datasets/full-time_employ_rate.csv',index_col='Country')

fte.head()

```
